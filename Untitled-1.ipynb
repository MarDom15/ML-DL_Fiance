```python
import yfinance as yf
import os

# Configuration de l'actif et de la période
ticker = "BTC-USD"  # Bitcoin en dollars US
start_date = "2020-01-01"  # Date de début
end_date = "2024-12-06"    # Date de fin

# Télécharger les données depuis Yahoo Finance
data = yf.download(ticker, start=start_date, end=end_date)

# Spécifiez le chemin vers le dossier local du repository
repo_path = r"C:\Users\marti\Desktop\Nouveau dossier\ML-DL_Fiance\BTC_USD_yahoofinace"
output_file = os.path.join(repo_path, "bitcoin_data.csv")

# Vérifier si le dossier existe
if not os.path.exists(repo_path):
    os.makedirs(repo_path)

# Enregistrer les données dans le fichier CSV
data.to_csv(output_file)
print(f"Les données ont été enregistrées dans le fichier : {output_file}")

```

```python
# Si la date est dans l'index, nous n'avons pas besoin de la convertir en colonne
# Si la date est une colonne, nous la convertissons en datetime
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])
else:
    data.index = pd.to_datetime(data.index)
```

```python
# Visualisation de la distribution des prix de clôture (Close)
plt.figure(figsize=(10, 6))
sns.histplot(data['Close'], kde=True, bins=30, color='blue')
plt.title('Distribution des prix de clôture (Close)')
plt.xlabel('Prix de clôture (Close)')
plt.ylabel('Fréquence')
plt.show()
```

```python
# Visualiser les variations au fil du temps (Close vs Date)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Prix de clôture')  # Utilisation de l'index si la date est dans l'index
plt.title('Évolution des prix de clôture dans le temps')
plt.xlabel('Date')
plt.ylabel('Prix de clôture (Close)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Détection des valeurs extrêmes (outliers) via boîte à moustaches
plt.figure(figsize=(8, 5))
sns.boxplot(data['Close'], color='green')
plt.title('Détection des outliers dans les prix de clôture (Close)')
plt.xlabel('Prix de clôture (Close)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

# Charger les données
data = pd.read_csv('BTC_USD_yahoofinance.csv')

# Utiliser 'Close' comme variable cible et 'Volume' comme caractéristique
X = data[['Volume']]  # Vous pouvez ajouter d'autres colonnes comme 'Open', 'High', 'Low'
y = data['Close']

# Diviser les données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Créer et entraîner le modèle
model = LinearRegression()
model.fit(X_train, y_train)

# Prédictions
y_pred = model.predict(X_test)

# Évaluation du modèle
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
